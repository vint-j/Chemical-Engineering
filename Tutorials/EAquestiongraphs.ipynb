```python
#T1EA2
import numpy as np
def func(z,P):
    R,T=8.3145e-2,182
    v=z*R*T/P*1000
    return v
Pval=np.array([30.58,24.74,18.82,13.67,9.62,6.62])
zval=np.array([0.652,0.745,0.819,0.875,0.915,0.943])
vval = func(zval,Pval)
vval
```

```python
import matplotlib.pylab as plt

linear_fit = np.polyfit(vval, zval, 1)
linear_func = np.poly1d(linear_fit)

quadratic_fit = np.polyfit(vval, zval, 2)
quadratic_func = np.poly1d(quadratic_fit)


v_range = np.linspace(min(vval), max(vval), 100)
plt.plot(vval, zval, 'o', label='Data points')
plt.plot(v_range, linear_func(v_range), label='Linear fit')
plt.plot(v_range, quadratic_func(v_range), label='Quadratic fit')

plt.xlabel('v val (mL)')
plt.ylabel('Z val')
plt.legend()
plt.show()

print("Linear fit equation:", linear_func)
print("Quadratic fit equation:", quadratic_func)
```
